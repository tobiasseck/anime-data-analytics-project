```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
main_df = pd.read_csv("../data/animes_better.csv")
main_df
```

```python
columns_to_drop = ['title', 'aired', 'episodes', 'members', 'ranked', 'score', 'link', 'img_url']
main_df = main_df.drop(columns=columns_to_drop)
main_df
```

```python
main_df['Producers'] = main_df['Producers'].apply(eval)
main_df['Studios'] = main_df['Studios'].apply(eval)
main_df['Themes'] = main_df['Themes'].apply(eval)
main_df['Streaming Platforms'] = main_df['Streaming Platforms'].apply(eval)
```

```python
def plot_correlation_with_popularity(column, title):
    exploded = main_df.explode(column)

    correlation = exploded.groupby(column)['popularity'].mean().sort_values(ascending=False)

    plt.figure(figsize=(16, 8))
    sns.barplot(y=correlation.index, x=correlation.values)
    plt.title(f'Top {title} Correlated with Popularity')
    plt.xlabel('Popularity')
    plt.ylabel(title)
    plt.show()

plot_correlation_with_popularity('Type', 'Types')
plot_correlation_with_popularity('Producers', 'Producers')
plot_correlation_with_popularity('Studios', 'Studios')
plot_correlation_with_popularity('Source', 'Sources')
plot_correlation_with_popularity('Themes', 'Themes')
plot_correlation_with_popularity('Demographic', 'Demographics')
plot_correlation_with_popularity('Streaming Platforms', 'Streaming Platforms')
```
